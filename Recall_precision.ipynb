```python
#Without cluster 1 second data
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

#data = np.loadtxt('data.txt')
data_rating =  pd.read_csv('All_without_cluster_1sec.csv', header=None)
data_rating=np.array(data_rating)
#print(data_rating)

Ground_trute =  pd.read_csv('GT_Label_1_20.csv', header=None)
Ground_trute=np.array(Ground_trute)

print("Without cluster")


for i in range(0, len(Ground_trute)):
    print("................",(i+1))
    actual = Ground_trute[i]
    predicted = data_rating[i]
    print(actual)
    print(predicted)
    results = confusion_matrix(actual, predicted) 
  
    print ('Confusion Matrix :')
    print(results) 
    print ('Accuracy Score :',accuracy_score(actual, predicted) )
    print ('Report : ')
    print (classification_report(actual, predicted)) 

   
```
